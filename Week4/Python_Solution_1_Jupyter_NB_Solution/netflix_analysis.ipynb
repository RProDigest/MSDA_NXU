##### From the analysis above, we observed that the columns director, cast, country, date_added, and rating have missing values. The counts of missing values in these columns are **1969, 570, 476, 11, and 10, respectively**.

##### To address these missing values, we have several options:

##### **1. Impute the Missing Values:** We can fill in the missing values using appropriate methods, such as using the mean, median, mode, or forward fill/backward fill techniques.
##### **2. Remove Rows with Missing Values:** We can remove the rows that contain missing values. This is useful if the missing values are relatively few and do not significantly impact the analysis.
##### **3. Do Nothing:** If the columns with missing values are not relevant to the visualizations or analyses we are interested in, we may choose to ignore the missing values.

```python
## In this first example, we ignore the missing values as we pick columns that dont have missing values, these include 'ratings', 'listed_in' and 'release_year'

sns.set_theme(palette="viridis") # Apply viridis a color blind friendly palette to my plots

# Example visualization: Ratings distribution
plt.figure(figsize=(10, 6))
sns.histplot(Netflix_shows_movies['rating'], kde=True, bins=20)
plt.title('Ratings Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()

# Count the occurrences of each genre
genre_counts = Netflix_shows_movies['listed_in'].str.split(',').explode().value_counts().head(10).reset_index()
genre_counts.columns = ['genre', 'count']

plt.figure(figsize=(12, 8))
   # sns.histplot(Netflix_shows_movies['listed_in'], kde=True)
sns.barplot(data=genre_counts, x='count', y='genre')
plt.title('Top 10 Most Watched Genres')
plt.xlabel('Number of Shows/Movies')
plt.ylabel('Genre')
plt.show()



# Example visualization: Another  column, such as 'release_year'
plt.figure(figsize=(10, 6))
sns.histplot(Netflix_shows_movies['release_year'], kde=True, bins=20)
plt.title('Release Year Distribution')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.show()

```

#### 1. The ratings chart above shows a bimodal distribution with the TV-MAT and TV-14 categories possesing the highest densities. This means Netflix can focus on these generes to maximize revenues.
#### 2. The distributions also show ratings that will need boosting using different techniques such as cross selling etc. In otherwords, Netflix, with additional data sources can formulate strategies to improve ratings in low performing genres.

#### 3. The Most watched genre chart shows international movies have the highest viewership and as such Netflix can focus in this segment to boost revenue share

#### 4. The release year distribution is left-skewed, with the peak occurring just before 2020. This likely suggests that the maturity of computer-generated imagery and the proliferation of computing power have democratized movie-making in recent years, boosting film production.

```python
# In the next example, we fill the  the rows(observations) that have missing values will forward filling from pandas and focus on plotting the director distribution
# Thus, Lets handle missing values with forward fill
Netflix_shows_movies.ffill(inplace=True)  # Fill missing values with forward fill

# Display the dataset's info after forward fill
info_ff = Netflix_shows_movies.info()

print(info_ff)

```

```python
# Plot the distribution of the top ten directors

import seaborn as sns
import matplotlib.pyplot as plt

# Count the occurrences of each director
director_counts = Netflix_shows_movies['director'].value_counts().head(10).reset_index()
director_counts.columns = ['director', 'count']

# Plot the distribution of the top directors
plt.figure(figsize=(12, 8))
sns.barplot(data=director_counts, x='count', y='director', hue='director', dodge=False, palette="viridis", legend=False)
plt.title('Top 10 Directors Distribution')
plt.xlabel('Number of Shows/Movies')
plt.ylabel('Director')
plt.show()
```

#### The bar graph above shows Eli Roth is is the top director followed by Jerry Seinfield
